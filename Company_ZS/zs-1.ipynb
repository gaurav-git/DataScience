```python
columns = list(set(data.columns)-set(["Disease Flag","Resident ID" ]))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data[columns], data["Disease Flag"], test_size=0.3, random_state=42, stratify=data["Disease Flag"])
```

```python
pd.Series(Y_test).value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, Y_train)
print("Features importance: ", rfc.feature_importances_)
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score
print("accuracy_score : ", accuracy_score(Y_test, rfc.predict(X_test)))
print("confusion matrix : ")
confusion_matrix(Y_test, rfc.predict(X_test))
```

```python
prob = rfc.predict_proba(X_test)
pred2 = np.asarray(prob[:,1]>0.1, np.int)
confusion_matrix(Y_test, pred2)
```

```python
roc_auc_score(y_true=Y_test, y_score=prob[:,1])
```
